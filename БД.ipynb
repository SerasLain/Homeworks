# Базы данных

Базы данных - это библиотеки с данными. Системы управления базами данных - библиотеки для управления этими библиотеками.
БД:
1. Реляционные - таблицы, где по столбцам и строкам информация
2. Нереляционные - информация хранится не в таблицах, а в чем-то другом (например, в json'е)

Реляционные СУБД: MySQL, PgSQL, SQL, SQLife, Access
Различаются возможностями, но делают примерно одно и то же. Бывают файловые, бывают нет.
Общение с базой данных происходит на SQL. Декларативный, в отличие от питона: то есть мы говорим, что хотим сделать, а дальше оно как-то само.

    Все команды SQL делятся на две части: DDL (управлять структурой БД - создавать/удалять таблицы) и DML (управлять данными - добавлять новые записи, удалять и т. д.)
    Команды - DDL: (Creat, ALTER, DROP), DMM: ISNSERT, SELECT, UPDATE, DELEATE
 
 Будем писать SQL запросы в тренировочную системы.
 На второй паре - создать реальную БД и использовать для хранения на веб-сервисе.
 
SELECT <name, address> FROM <таблица, owners> WHERE <условие, ID > 1>
Результат - подмножество таблицы или нескольких таблиц.
В сортировке можно адать DESC - обратный порядок.
